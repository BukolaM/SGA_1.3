```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Reading the file using pandas
df = pd.read_csv('./company_sales_data.csv')
df
```

## Making a Copy of The Dataset
I would be working with this copy through while the original dataframe would be for referencing when needed.

```python
#making a copy of my dataset
df_sales = df.copy()
```

```python
df_sales
```

```python
df_sales.plot(x='month_number', y=['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer'], figsize = (15,5))
plt.legend(loc = 'upper left')
plt.title("comparison of company's products sales by months")
plt.xlabel('months')
plt.ylabel("total sales of different company's products")
plt.show()
```
