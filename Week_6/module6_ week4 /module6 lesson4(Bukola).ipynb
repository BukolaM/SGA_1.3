### <span style='color:blue'> Question 2.   An inner merge operation between the orders table and products table. Use the resulting table to perform another inner merge operation with the customers table. State your observations in a text cell.</span>

```python
#creating an inner merge operation between the orders table and products table
df_ordersProducts = df_orders.merge(df_products, how='inner', on='product_id')
df_ordersProducts
```

#### From the df_ordersProducts table above, it can be seen that df_ordersProducts returns a dataframe with only those rows that have common properties. 
I observed that the only rows contained in the df_ordersProducts(resulting DataFrame) are the ones where the `product_id` appears in both `df_orders` and `df_products` DataFrames.

```python
# using the resulting table to perform another inner merge operation with the customers table
df_all = df_ordersProducts.merge(df_customers, how='inner', on='customer_id')
df_all
```

#### From the  df_all table above, it can be seen that df_all returns a dataframe with only those rows that have common characteristics. 

I observed that only the rows contained in the df_all(resulting DataFrame) are the ones where the `customer_id` exist in both `df_ordersProducts(resulting table)` and `df_all` DataFrames.

### <span style='color:blue'> Question 3.  Perform an inner merge operation between customers and orders tables. Here you need to specify the columns you’ll be merging on .</span>

```python
##In this, I would be merging on customer_id
df_customersOrders = df_customers.merge(df_orders, how='inner', on='customer_id')
df_customersOrders
```

### <span style='color:blue'> Question 4.  Perform an outer merge between customers and products tables. (see the .merge argument ‘how’). State your observations in a text cell. </span>

```python
df_customersProducts = pd.merge(df_customers,df_products,left_index=True,right_index=True,how = 'outer')
df_customersProducts
```

#### From the df_customersOrders table above, it can be seen that df_customersProducts returns a dataframe of all rows from both DataFrames
I observed that both the customer and the product dataframe had no matching row. However, I was able to join both dataframes using the matching index of both DataFrames.

After the merge, I also observed that the customer table has NAN value. This can be attributed to the fact that the product table has more rows than the customer dataframe.

### <span style='color:blue'> Question 5. Perform a left merge between orders and customers tables. State your observations in a text cell. </span>

```python
##In this, I would be merging on customer_id
df_ordersCustomers = df_orders.merge(df_customers, how='left', on='customer_id')
df_ordersCustomers
```

#### From the df_customersOrders table above, it can be seen that I aligned both tables using the customer_id.

After the merge, I observed that the merged DataFrame returned NaNs in some of its columns. This is due to the fact that  the missing values didn’t have corresponding data points in the customer table.


### <span style='color:blue'> Question 6. Perform a right merge between product and customers tables. State your observations in a text cell. </span>

```python
df_productsCustomers = pd.merge(df_products, df_customers,left_index=True,right_index=True,how = 'right')
df_productsCustomers
```

I was able to join both dataframes using the matching index of both DataFrames.

After the merge, I observed that both the customer and the product only brings out the matching rows. This can be attributed to the fact that the product table has more rows than the customer dataframe. However, the resulting DataFrame only returns all the rows on the right side product table and match the rows on the left side of the `customers`.
