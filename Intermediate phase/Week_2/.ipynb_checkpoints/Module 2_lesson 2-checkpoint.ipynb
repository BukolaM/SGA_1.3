```python
#importing the required libraies
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
# establish connection with the database
engine = create_engine('mysql+pymysql://root:@localhost:3306/demo', pool_recycle=3600)
```

```python
# read table data using sql query
df = pd.read_sql('SELECT * FROM udemy_courses', con = engine)
df.head(2)
```

### 1. Write a query for paid courses, showing the  course with the highest number of subscribers per month

```python
#selecting the individual courses based on all their months
# i also extracted the month from using sql format from the datetime in the csv
df = pd.read_sql("SELECT subject, DATE_FORMAT(`published_timestamp`,'%%M') as datemonth, \
SUM(num_subscribers) as subcribers \
FROM demo.udemy_courses \
GROUP BY subject, datemonth", con = engine)
df
```

```python
#ordering the course with the highest number of subscribers per month
df_monthlySubscribers= df.loc[df.groupby('datemonth', sort=False).subcribers.idxmax()]
df_monthlySubscribers
```

- This shows that Web Development courses(out of all the courses) has the highest amount of subscribers in all the months

### 2.  Write a query for paid courses, showing the course with the highest sales per month

```python
#selecting the individual courses based on all their months
#based on my dataset, i'm considering where price = 'true'
# i also extracted the month from using sql format from the datetime in the csv

df = pd.read_sql("SELECT subject, DATE_FORMAT(`published_timestamp`,'%%M') as datemonth, \
SUM(price) as sales \
FROM demo.udemy_courses \
WHERE is_paid= 'TRUE' \
GROUP BY subject, datemonth", con = engine)
df
```

```python
#ordering the course with the highest number of sales per month
df_monthlySales = df.loc[df.groupby('datemonth', sort=False).sales.idxmax()]
df_monthlySales
```

- We can see that both `Business Finance` and `Web Development` courses has the highest number of sales per month
