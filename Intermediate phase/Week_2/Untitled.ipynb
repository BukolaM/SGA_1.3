## 1. Write a query for paid courses, showing the  course with the highest number of subscribers per month

```python
#extraction and conversion analysis of published_timestamp and converting to python datetime
df['published_date'] = pd.to_datetime(df.published_timestamp).dt.strftime('%Y-%m-%d')

#conversion to day only
df['published_day'] = pd.to_datetime(df.published_timestamp).dt.day

#conversion to month only
df['published_month'] = pd.to_datetime(df.published_timestamp).dt.strftime('%B')

#conversion to weekdays only
df['published_day_of_week'] = pd.to_datetime(df.published_timestamp).dt.day_name()
```

```python
df = pd.read_sql("SELECT subject,\
                 SUM(num_subscribers),\
                 DATE_FORMAT(`published_timestamp`,'%%M') AS showdate\
                 FROM udemy_courses_1\
                 where is_paid = 'TRUE'\
                 GROUP BY subject",con = engine)

df
```
