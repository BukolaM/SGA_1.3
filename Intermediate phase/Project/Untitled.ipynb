```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
from datetime import datetime
```

```python
#Importing the datetime to calculate the time for scraping the 200000 tweets
start_time = datetime.now()

loc = '9.0820, 8.6753, 923768km'
Ng = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'geocode:"{}" since:2022-12-15 until:2022-12-16'.format(loc)).get_items(), 15000))

end_time = datetime.now()

Ng
print('Duration: {}'.format(end_time - start_time))
```
