## Creating a Copy of The Dataset

## Data Cleaning

## <span style='color: blue'> Focus on 3 or more Cleaning

#### Issue 1: Drop all rows with missing values
- Here, I would be dropping the rows with null values as they are non significant to this dataset

```python
#deleting all rows of NaN in the dataset
df_cancer1 = df_cancer1.dropna(axis = 0, how ='any', thresh = None, subset = None, inplace=False)

# resetting the index
df_cancer1 = df_cancer1.reset_index(drop = True)
```

#### Issue 2: Drop all duplicates rows 
- There are currently 5 rows that are duplicates, so I would be dropping them to know the exact number of unique rows we have.

```python
#viewing the duplicates rows
df_cancer1[df_cancer1.duplicated()]
```

```python
#there are currently 5 duplicates rows
df_cancer1.duplicated().sum()
```

```python
#dropping the duplicate rows
df_cancer1.drop_duplicates(inplace=True)
```

```python
#confirming there are no duplicates rows
df_cancer1.duplicated().sum()
```

#### Issue 3: Dropping Columns
- Unnecessary column `id` would be dropped off. This `id` column is inconsistent  and therefore would not be required for further analysis

```python
#dropping the column
df_cancer1.drop(['id'], axis = 1, inplace = True)
```

## Exporting the Cleaned Data to a New CSV

```python
#exporting the cleaned data
df_cancer = df_cancer1.to_csv('cancer_data_clean.csv', index = False)
```

## Exploratory Data Analysis

## <span style='color: blue'>  Focus on 3 or more explorations at EDA stage

#### Question 1. What is the distribution of cancer diagnosis?

```python
#grouping based on diagnosis
df_diagnosis =  df_cancer['diagnosis'].groupby(df_cancer['diagnosis']).size().reset_index(name='counts')
df_diagnosis
```

```python
#function to set titles and labels
def titles_and_labels(plt, ax, title, x_title, y_title, fontsize_title, fontsize_xlabel, fontsize_ylabel):
    ax.grid(False)
    ax.set_title(title, fontsize = fontsize_title)
    ax.set_xlabel(x_title, fontsize = fontsize_xlabel)
    ax.set_ylabel(y_title, fontsize = fontsize_ylabel)
    plt.show()
```

```python
# plotting the results
plt.figure(figsize=(10,5))
ax = sns.barplot(x = 'diagnosis', y = 'counts', data = df_diagnosis, palette = 'tab20')

#setting title and labels
titles_and_labels(plt, ax, 'distribution of cancer diagnosis', 'cancer diagnosis', 'counts', 10, 10, 10 )
```

- The chart above shows the distribution of cancer diagnosis in the dataset

#### Question 2. What is the comparison of radius mean, area mean, and concativity mean of cancer diagnosis stages?

```python
#plotting the chart
fig = plt.figure(figsize=(8,8))
ax = sns.boxplot(y='radius_mean', x='diagnosis',  data=df_cancer, palette='mako')

#setting title and labels
titles_and_labels(plt, ax, 'radius mean of cancer stages',
                  'cancer diagnosis', 'radius_mean', 10, 10, 10)
```

- The box above shows the `radius mean` quartiles of the malignant stage while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers.


```python
#plotting the chart
fig = plt.figure(figsize=(8,8))
ax =  sns.violinplot(y='area_mean', x='diagnosis',  data=df_cancer, palette='coolwarm')

#setting title and labels
titles_and_labels(plt, ax, 'area mean of cancer stages',
                  'cancer diagnosis', 'area_mean', 10, 10, 10)
```

- The box above shows the `area mean` quartiles of the malignant stage while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers.

```python
#plotting the chart
fig = plt.figure(figsize=(8,8))
ax = sns.boxplot(y='concavity_mean', x='diagnosis',  data = df_cancer, palette='viridis')

#setting title and labels
titles_and_labels(plt, ax, 'radius mean of cancer stages',
                  'cancer diagnosis', 'concavity_mean', 10, 10, 10)
```

- The box above shows the `concativity mean` quartiles of the malignant stage while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers.

#### Question 3. Plot and customize a correlations heatmap for the dataset

```python
fig = plt.figure(figsize=(15,15))
sns.heatmap(df_cancer.corr(),linecolor='white', annot= True, linewidths = 2, fmt= '.1f', cmap="vlag")
plt.title('correlations heatmap of the dataset', fontsize= 12)
plt.show()
```

- From the correlation heatmap above, we can see that our dataset have too many correlation of one feature to each other.
